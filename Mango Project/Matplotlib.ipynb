# Line plot - plt.plot()

## single plot - 1 graph

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv(mango_file_path)
weight=data["WEIGHT_g"]
shortW=data["DIA1_mm_shortwidth"]
longW=data["DIA2_mm_longwidth"]
distance=data["Distance"]
plotNames=["weight","shortW","longW","distance"]
plt.plot(weight)
plt.plot(shortW)
plt.plot(longW)
plt.plot(distance)
plt.legend(plotNames)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("XY Graph")
plt.xlim(0,250)
plt.ylim(0,500)
plt.plot(weight,marker='o',ms=10,mec='r',mfc='c',linestyle='dotted',linewidth=1)
plt.show()
```

# Subplot - plt.subplot()

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the data
data = pd.read_csv(mango_file_path)
weight = data["WEIGHT_g"]
shortW = data["DIA1_mm_shortwidth"]
longW = data["DIA2_mm_longwidth"]
distance = data["Distance"]

# Colors for the bars
clr = ['r', 'b', 'g', 'y', 'c']

# Create a bar plot for the first 5 values of weight
b = plt.bar(range(len(weight[:5])), weight[:5], alpha=0.5, width=0.3)

# Set the color for each bar
for cnt, bar in enumerate(b):
    bar.set_color(clr[cnt])  # Set color of each bar
    # Place text above each bar
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), 
             str(weight[cnt]), ha='center', va='bottom')

# Labels and title
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("XY Graph")

# Set the y-axis limits
plt.ylim(0, max(weight[:5]) + 50)

# Add grid lines
plt.grid(axis='x')
plt.grid(axis='y')

# Plot a dotted line connecting the weight values
plt.plot(weight[:5], marker='o', ms=10, mec='r', mfc='c', linestyle='dotted', linewidth=1)

# Customize ticks on the y and x axes
plt.yticks(range(0, 301, 25))
plt.xticks(range(0, 6, 1))

# Show the plot
plt.show()
```

# horizontal bar plot - plt.barh()

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv(mango_file_path)
weight=data["WEIGHT_g"]
shortW=data["DIA1_mm_shortwidth"]
longW=data["DIA2_mm_longwidth"]
distance=data["Distance"]
plotNames=["weight","shortW","longW","distance"]
clr=['r','b','g','y','c']
b=plt.barh(range(len(weight[:5])),weight[:5],alpha=0.5,height=0.2)
cnt=0
for i in clr:
    b[cnt].set_color(i)
    cnt=cnt+1
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("XY Graph")
plt.legend(b,clr)
plt.grid(axis='x')
plt.grid(axis='y')
plt.xlim(0,max(weight[:5])+50)
plt.xticks(range(0,301,25))
plt.yticks(range(0,6,1))
plt.show()
```

# pie chart - plt.pie()

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv(mango_file_path)
weight=data["WEIGHT_g"]
shortW=data["DIA1_mm_shortwidth"]
longW=data["DIA2_mm_longwidth"]
distance=data["Distance"]
plotNames=["weight","shortW","longW","distance"]
clr=['r','b','g','y','c']
plt.pie(weight[:5],labels=clr,startangle=0)
plt.legend(clr)
plt.show()
```

# Scatter plot - plt.scatter

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv(mango_file_path)
weight=data["WEIGHT_g"]
shortW=data["DIA1_mm_shortwidth"]
longW=data["DIA2_mm_longwidth"]
distance=data["Distance"]
plotNames=["weight","shortW","longW","distance"]
clr=['r','b','g','y','c']
plt.scatter(range(len(weight)),weight,label="Weight")
plt.scatter(range(len(shortW)),shortW,marker='o',s=100,c='y',edgecolors='r',facecolors='k',linestyle=':',label="ShortWidth")
plt.scatter(range(len(longW)),longW,label="LongWidth")
plt.scatter(range(len(distance)),distance,label="Distance")
plt.grid(axis='x')
plt.grid(axis='y')
plt.ylim(0,450)
plt.yticks(range(0,450,20))
plt.xlim(0,250)
plt.xticks(range(0,250,25))
plt.legend()
plt.show()
```
