```python
'''
Feature selection of independent and dependent variables
'''

x=datatest_1[['Year', 'Crash Descriptor', 'Day of Week','Police Report', 'Collision Type Descriptor', 'Traffic Control Device','Road Surface Conditions']]
y=datatest_1["Number of Vehicles Involved"]
```

```python
# Loading and importing packages for traing , building a model and evaluation metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,accuracy_score,precision_score, recall_score,f1_score
import numpy as np      # For array operations , reshaping an array

model=LinearRegression()                                            # Model selection
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)   # Train , Test, Split data
print("x_train shape : ",np.array(x_train.shape))                   # To get shape of Traing dataset - x_train
print("y_train shape : ",np.array(y_train.shape))                   # To get shape of Traing dataset - y_train
print("x_test shape : ",np.array(x_test.shape))                     # To get shape of Testing dataset - x_test
print("y_test shape : ",np.array(y_test.shape))                     # To get shape of Testing dataset - y_test

model.fit(x_train,y_train)                                          # To train the model with trainig dataset
y_pred=model.predict(x_test)                                        # Predict the data with testing dataset

print("Training score : ",model.score(x_train,y_train))                                     # To get training score
print("Testing score : ",model.score(x_test,y_test))                                        # To get testing score
print("Model score : ",model.score(x_test,y_test))                                          # To get model score
print("Model score (R-squared): ", model.score(x_test, y_test))                             # To get r2 score
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))                      # To get mean squared error
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))       # To get root mean squared error
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))                    # To get mean absolute error
```

```python
# Import classification models from scikit-learn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB, BernoulliNB
import numpy as np      # For array operations , reshaping an array
                                    

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)   # Train , Test, Split data
print("x_train shape : ",np.array(x_train.shape))                   # To get shape of Traing dataset - x_train
print("y_train shape : ",np.array(y_train.shape))                   # To get shape of Traing dataset - y_train
print("x_test shape : ",np.array(x_test.shape))                     # To get shape of Testing dataset - x_test
print("y_test shape : ",np.array(y_test.shape))                     # To get shape of Testing dataset - y_test

Knn_model=KNeighborsClassifier()                                        # Model selection for KNeighbour
print("*"*20,"KNeighborsClassifier","*"*20)
Knn_model.fit(x_train,y_train)                                          # To train the model with trainig dataset
y_pred=Knn_model.predict(x_test)                                        # Predict the data with testing dataset
print("Training score : ",Knn_model.score(x_train,y_train))                                 # To get training score
print("Testing score : ",Knn_model.score(x_test,y_test))                                    # To get testing score
print("Model score : ",Knn_model.score(x_test,y_test))                                      # To get model score
print("Model score (R-squared): ", Knn_model.score(x_test, y_test))                         # To get r2 score
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))                      # To get mean squared error
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))       # To get root mean squared error
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))                    # To get mean absolute error

DT_model=DecisionTreeClassifier()
print("*"*20,"DecisionTreeClassifier","*"*20)
DT_model.fit(x_train,y_train)                                          # To train the model with trainig dataset
y_pred=DT_model.predict(x_test)                                        # Predict the data with testing dataset
print("Training score : ",DT_model.score(x_train,y_train))                                 # To get training score
print("Testing score : ",DT_model.score(x_test,y_test))                                    # To get testing score
print("Model score : ",DT_model.score(x_test,y_test))                                      # To get model score
print("Model score (R-squared): ", DT_model.score(x_test, y_test))                         # To get r2 score
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))                      # To get mean squared error
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))       # To get root mean squared error
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))                    # To get mean absolute error

RF_model=RandomForestClassifier()
print("*"*20,"RandomForestClassifier","*"*20)
RF_model.fit(x_train,y_train)                                          # To train the model with trainig dataset
y_pred=RF_model.predict(x_test)                                        # Predict the data with testing dataset
print("Training score : ",RF_model.score(x_train,y_train))                                 # To get training score
print("Testing score : ",RF_model.score(x_test,y_test))                                    # To get testing score
print("Model score : ",RF_model.score(x_test,y_test))                                      # To get model score
print("Model score (R-squared): ", RF_model.score(x_test, y_test))                         # To get r2 score
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))                      # To get mean squared error
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))       # To get root mean squared error
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))                    # To get mean absolute error

svc_model=SVC()
print("*"*20,"SVC","*"*20)
svc_model.fit(x_train,y_train)                                          # To train the model with trainig dataset
y_pred=svc_model.predict(x_test)                                        # Predict the data with testing dataset
print("Training score : ",svc_model.score(x_train,y_train))                                 # To get training score
print("Testing score : ",svc_model.score(x_test,y_test))                                    # To get testing score
print("Model score : ",svc_model.score(x_test,y_test))                                      # To get model score
print("Model score (R-squared): ", svc_model.score(x_test, y_test))                         # To get r2 score
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))                      # To get mean squared error
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))       # To get root mean squared error
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))                    # To get mean absolute error


GNB_model=GaussianNB()
print("*"*20,"GaussianNB","*"*20)
GNB_model.fit(x_train,y_train)                                          # To train the model with trainig dataset
y_pred=GNB_model.predict(x_test)                                        # Predict the data with testing dataset
print("Training score : ",GNB_model.score(x_train,y_train))                                 # To get training score
print("Testing score : ",GNB_model.score(x_test,y_test))                                    # To get testing score
print("Model score : ",GNB_model.score(x_test,y_test))                                      # To get model score
print("Model score (R-squared): ", GNB_model.score(x_test, y_test))                         # To get r2 score
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))                      # To get mean squared error
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))       # To get root mean squared error
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))                    # To get mean absolute error


BNB_model=BernoulliNB()
print("*"*20,"BernoulliNB","*"*20)
BNB_model.fit(x_train,y_train)                                          # To train the model with trainig dataset
y_pred=BNB_model.predict(x_test)                                        # Predict the data with testing dataset
print("Training score : ",BNB_model.score(x_train,y_train))                                 # To get training score
print("Testing score : ",BNB_model.score(x_test,y_test))                                    # To get testing score
print("Model score : ",BNB_model.score(x_test,y_test))                                      # To get model score
print("Model score (R-squared): ", BNB_model.score(x_test, y_test))                         # To get r2 score
print("Mean Squared Error (MSE):", mean_squared_error(y_test, y_pred))                      # To get mean squared error
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, y_pred)))       # To get root mean squared error
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, y_pred))                    # To get mean absolute error
```
