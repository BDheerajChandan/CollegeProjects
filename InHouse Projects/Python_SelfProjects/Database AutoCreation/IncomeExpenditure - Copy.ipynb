# 5. Organise and extract required data from Excel sheet (Number of transactions)

### Display multiple transaction data

```python
# 5.1 To find number of multiple transactions (Multiple transaction data)
import pandas as pd
df3=pd.read_excel(excel_path)
dt=list(df3[df3.columns[2]])
c=0
for i in dt:
    cnt=dt.count(i)
    if cnt>1:
        print("On Date: ",i," Number of multiple Transactions done: ",cnt)
        c=c+1
        dt=dt[dt.index(i)+cnt:]
print("\nNumber of Multiple transactions : ",c)
```

### Display single transaction data

```python
# 5.2 To find number of Single transactions (Single transaction data)
import pandas as pd
df3=pd.read_excel(excel_path)
dt=list(df3[df3.columns[2]])
c=0
for i in dt:
    cnt=dt.count(i)
    if cnt==1:
        print("On Date: ",i," Number of single Transactions done: ",cnt)
        c=c+1
        dt=dt[dt.index(i)+cnt:]
print("\nNumber of single transactions : ",c)
```

### Display full transaction data

```python
# 5.3 To find number of transactions done (transactions on each day)
import pandas as pd
df3=pd.read_excel(excel_path)
dt=list(df3[df3.columns[2]])
c=0
multi_trans=0
single_trans=0
for i in dt:
    cnt=dt.count(i)
    if cnt>1:
        print("On Date: ",i," Number of transactions done: ",cnt)
        c=c+1
        multi_trans=multi_trans+1
        dt=dt[dt.index(i)+cnt:]
    elif cnt==1:
        print("On Date: ",i," Number of transactions done: ",cnt)
        c=c+1
        single_trans=single_trans+1
    else:
        pass
print("\nNumber of Days collected : ",c)
print("Number of Multiple transactions done on single day : ",multi_trans)
print("Number of Single transactions done on single day : ",single_trans)
```

# 6. SQL using Python - Create,Read & Write

### Read data from excel and create database db file

```python
# 6.1 Read data from excel and create database db file at specific location.
import pandas as pd
import sqlite3
df4=pd.read_excel(excel_path)
connection = sqlite3.connect(sql_path)
```

### Write Excel data to Database file

```python
# 6.2 Create a table in database and write data after reading data from excel
crsr = connection.cursor()
def database():
    #crsr = connection.cursor()
    sql_command = f"""CREATE TABLE IncExpData (
                        '{df4.columns[1]}' INTEGER,
                        '{df4.columns[2]}' DATE,
                        '{df4.columns[3]}' VARCHAR(20),
                        '{df4.columns[4]}' VARCHAR(20),
                        '{df4.columns[5]}' FLOAT,
                        '{df4.columns[6]}' VARCHAR(30),
                        '{df4.columns[7]}' VARCHAR(30),
                        '{df4.columns[8]}' FLOAT,
                        '{df4.columns[9]}' FLOAT,
                        '{df4.columns[10]}' FLOAT
                    );"""
    crsr.execute(sql_command)
    for i in range(len(df4)):
        crsr.execute(f"""INSERT INTO IncExpData VALUES ('{df4[df4.columns[1]][i]}','{df4[df4.columns[2]][i]}','{df4[df4.columns[3]][i]}',
                                                      '{df4[df4.columns[4]][i]}','{df4[df4.columns[5]][i]}','{df4[df4.columns[6]][i]}',
                                                      '{df4[df4.columns[7]][i]}','{df4[df4.columns[8]][i]}','{df4[df4.columns[9]][i]}',
                                                      '{df4[df4.columns[10]][i]}');
                    """)
    connection.commit()
try:
    database()
except:
    #crsr = connection.cursor()
    crsr.execute(f"DROP TABLE IncExpData")
    connection.commit()
    database()
```
